## Distributions in Python using libraries


```python
# Importing libraries

import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import uniform
```

```python
# Specifying the type of plots to be generated with Seaborn

sns.set(color_codes=True)
sns.set(rc={'figure.figsize':(5,5)})
```

### Uniform Distribution

```python
n = 1000
start = 10
width = 50

uniform_dist = uniform.rvs(size=n, loc=start, scale=width)
```

```python
ax = sns.distplot(uniform_dist,                                #creating object for histogram
                 bins = 100,                                   #number of bins in histogram
                 kde=True,                                     #kernel density
                 color='red',
                 hist_kws={"linewidth":15, 'alpha':1})         #line width
ax.set(xlabel='Uniform Distribution', ylabel='Frequency')      #setting labels
```

### Normal Distribution

```python
from scipy.stats import norm

size=10000
loc = 0                                         #loc = mean of 0
scale = 1                                       #scale = standard deviation of 1
normal_dist = norm.rvs(size, loc, scale)        #standard normal distribution
```

```python
ax2 = sns.distplot(normal_dist,
                  kde=True,
                  color='skyblue',
                  bins=100,
                  hist_kws={"linewidth":15, "alpha":1})

ax2.set(xlabel="Normal Distribution", ylabel="Frequency")
```

### Exponential Distribution

```python
from scipy.stats import expon

exponential_data = expon.rvs(loc=0, scale=1, size=1000)    #scale arguement = 1/lambda
```

```python
ax3 = sns.distplot(exponential_data,
                  kde=True,
                  color='skyblue',
                  bins=100,
                  hist_kws={"linewidth":15, "alpha":1})
ax3.set(xlabel='Exponential Distribution', ylabel="Frequency")
```

### Poisson Distribution

```python


from scipy.stats import poisson
poisson_data = poisson.rvs(mu=3, size=1000)
```

```python
ax4 = sns.distplot(poisson_data,
                  kde=True,
                  color='green',
                  bins=30,
                  hist_kws={"linewidth":15, "alpha":1})
ax4.set(xlabel='Poisson Distribution', ylabel="Frequency")
```

### Binomial Distribution

```python
from scipy.stats import binom

binomial_data = binom.rvs(size=100, n=10, p=0.8)
```

```python
ax5 = sns.distplot(binomial_data,
                  kde=True,
                  color='red',
                  
                  hist_kws={"linewidth":15, "alpha":1})
ax5.set(xlabel='Binomial Distribution', ylabel="Frequency")
```

### Bernoulli Distribution

```python
from scipy.stats import bernoulli
bernoulli_data = bernoulli.rvs(size=100, p=0.4)
```

```python
ax4 = sns.distplot(bernoulli_data,
                  kde=True,
                  color='green',
                  bins=30,
                  hist_kws={"linewidth":15, "alpha":1})
ax4.set(xlabel='Bernoulli Distribution', ylabel="Frequency")
```
